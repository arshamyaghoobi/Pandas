## Basic Analysis 

### value_counts(normalize=False,sort=True,ascending=False,bins=None,dropna=True) 

#### This method returns a series object, counting all the unique values. 
There are two things that need to be considered regarding value_counts method:

1- As this is returning a count of unique values, the first value is the most frequently occuring element and the second is the second most frequently occuring element and so on. This order can be reversed by just setting theh ascending flag to True.

2- Dropna, one of the parameters within the value_counts in True by default and you will not get a count of na values. The na values remember are the missing data values. If your dataset has the significant number of na values, this can be misleading and you can this feature off by setting dropna to False.

```python
oo.Edition.value_counts()
```

```python
oo.Gender.value_counts()
```

```python
oo.Gender.value_counts(ascending=True)
```

```python
oo.Athlete.value_counts()
```

## Types of Plot

### Plot()

#### By default the type of plot when you call plot method is "line" -----> plot(kind='line') 

### Types
### • plot(kind='line')

-- Best for tracking changes over a period of time.

### • plot(kind='bar')

-- Best for comparing between different groups.

### • plot(kind='barh')

-- The same as vertical bar graphs, horizantal bar graphs are the best for comparision between different groups.

### • plot(kind='pie')

-- Best for comparing parts of a whole, they don't show changes over the time.

### What were the different sports in the first olympics? PLot them using different graphs. 

```python
fo = oo[oo.Edition == 1896]
fo.head()
```

#### Line plot 

```python
fo.Sport.value_counts().plot(kind='line');
```

#### Bar plot 

```python
fo.Sport.value_counts().plot(kind='bar');
```

#### Horizontal Bar plot 

```python
fo.Sport.value_counts().plot(kind='barh');
```

#### Pie plot 

```python
fo.Sport.value_counts().plot(kind='pie')
```

### Plot Colors 

### plot(color='red') 

```python
fo.Sport.value_counts().plot(kind='bar',color='maroon');
```

### Plot Size

### figsize() 

```python
fo.Sport.value_counts().plot(kind='line',color='maroon',figsize=(10,4));
```

### Color Maps 

### Classes of colormaps 

#### Sequential 

-- The sequential should be used for representing information tha has ordering.

-- There is a change in lightness, often over a single hue.


#### Diverging 

-- Diverging is to be used when the information being plotted diviates around a middle value.

-- Here, there are often different colors being used.

#### Qualitative

-- The qualitative class is used to represent information which does not have any ordering or relationship.

-- Its of miscellaneous colors being used for this class.

```python
fo.Sport.value_counts().plot(kind='pie',colormap='Pastel1');
```
